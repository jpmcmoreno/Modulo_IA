Sabemos que la IA es una herramienta muy buena para resolver problemas asi que veremos si es capaz de resolver el problema abierto de la conjetura de collatz.  \\
Se le pregunto a una Chatgpt sobre este codigo.

```python
def collatz_conjecture(n):
    # inicializamos el conjunto con el primer número
    sequence = {n}

    while n != 1:
        if n % 2 == 0:
            # si es par, dividimos entre 2
            n = n // 2
        else:
            # si es impar, multiplicamos por 3 y sumamos 1
            n = 3 * n + 1
        # añadimos el siguiente número a la secuencia
        sequence.add(n)

    return sequence

# ejemplo de uso
n = 10
sequence = collatz_conjecture(n)
print(f"Conjetura de Collatz para {n}: {sequence}")
```

El codigo funciona sin embargo hay algunos detalles que se pueden mejorar como: \\

2. Que pregunta el input. \\

```python
def collatz_conjecture(n):
    # inicializamos la lista con el primer número
    sequence = [n]

    while n != 1:
        if n % 2 == 0:
            # si es par, dividimos entre 2
            n = n // 2
        else:
            # si es impar, multiplicamos por 3 y sumamos 1
            n = 3 * n + 1
        # añadimos el siguiente número a la secuencia
        sequence.append(n)

    return sequence

# pedimos al usuario el número de entrada
n = int(input("Introduce un número para la conjetura de Collatz: "))

# calculamos la secuencia y la mostramos en el orden de las iteraciones
sequence = collatz_conjecture(n)
print(f"Secuencia de la conjetura de Collatz para {n}:")
for number in sequence:
    print(number)
```

## ¿Puede la inteligencia artificial resolver el problema analíticamente?

**Input:** ¿Puedes resolver la conjetura de collatz?

**Output de Chatgpt:** Como modelo de lenguaje, no soy capaz de resolver problemas matemáticos por mi cuenta. Además, la conjetura de Collatz es un problema extremadamente difícil que ha resistido los esfuerzos de muchos matemáticos brillantes durante décadas.


**Input:** demuestra que si aplico el algoritmo 3n+1 termino en el 1

**Output:** El problema del 3n + 1, también conocido como el problema de Collatz, es un problema matemático sin resolver que afirma que para cualquier número entero positivo n, si se aplica repetidamente la siguiente regla:

Si n es par, divida n por 2.
Si n es impar, multiplique n por 3 y sume 1.
eventualmente llegará al número 1.

Para demostrar que el algoritmo siempre termina en 1, se puede demostrar por inducción.

Para el caso base, si n = 1, entonces el algoritmo termina inmediatamente en 1.

Ahora, asumamos que para algún entero positivo k, el algoritmo 3k + 1 siempre termina en 1. Consideremos el caso donde n = 2k.

En este caso, el algoritmo toma n y lo divide por 2, lo que da k. Por lo tanto, ahora estamos en la situación en la que 3k + 1 siempre termina en 1 (por la hipótesis de inducción), y hemos reducido el tamaño de n a la mitad.

En el caso donde n = 2k + 1, el algoritmo toma n, lo multiplica por 3 y le suma 1, lo que da 3(2k+1) + 1 = 6k + 4. Luego, dividimos este número por 2, lo que nos da 3k + 2. Ahora estamos en la misma situación que en el caso anterior, pero con un número más grande.

Por lo tanto, en ambos casos, hemos reducido el tamaño de n y estamos en una situación donde, por la hipótesis de inducción, el algoritmo 3n + 1 siempre termina en 1. Como resultado, podemos concluir que el algoritmo 3n + 1 siempre termina en 1 para cualquier número entero positivo n.







Puede que tenga sentido, aunque falta rigurosidad ya que no esta aplicando el metodo de induccion adeduadamente.

### Regresiones

Es un proceso estadistico para estomar relaciones entre variables dependientes (criterio) y una o mas variables independientes (predictores).

Un problema de regresión es cuando la variable de salida es un valor real o continuo, como “salario” o “peso”.
Se pueden usar muchos modelos diferentes, el más simple es la regresión lineal. Intenta ajustar los datos con la mejor recta que pasa por los puntos.

GeeksforGeeks. (2019). Types of Regression Techniques in ML. https://www.geeksforgeeks.org/types-of-regression-techniques/

#### Regresión Lineal
La regresión lineal es un tipo de algoritmo que calcula la relacion lineal de una variable dependiente y una o mas variables independientes.

Cuando el numero de variables independientes es uno, se conoce como regresión lineal de una variable, y en el caso donde haya más de una variable independiente se conoce como una regresión lineal multivariable.

Una regresión lineal se escribe:
$$y= ax+b$$
* $a$ se le dice peso
* $b$ se le dice sesgo

https://www.geeksforgeeks.org/ml-linear-regression/

#### Regresión Polinomica

De igual forma que una regresión lineal este intenta encontrar la relación de la variable dependiente con la independiente con un polinomio de grado n.

una regresión polinomica se escribe:
$$y=a+b_1x+b_2x^2+b_3x^3+\cdots+b_nx^n$$

la clasificación es el problema de identificar a cuál de un conjunto de categorías (subpoblaciones) pertenece una nueva observación, sobre la base de un conjunto de datos de entrenamiento que contiene observaciones y cuya pertenencia a categorías se conoce.

GeeksforGeeks. (2016). Getting started with Classification. https://www.geeksforgeeks.org/getting-started-with-classification/

#### k vecinos cercanos

#### Naive Bayes

Usa algoritmos basados en el teorema de bayes, como usan la probabilidad condicional los algoritmos mediran que tan probable es un suceso dado una condición.

#### Arbol de decisión

Un arbol de decisión es un diagrama de flujo con la estructura de un arbol, donde cada nodo interno denota una prueba en un atributo, cada rama representa un resultado de la prueba, y cada hoja (nodo terminal) contiene una etiqueta de clase.

##Técnicas de preprocesamiento de datos
Se usan tecnicas para preprocesar los datos para el algoritmo de machine learning.

**Datos de ejemplo**

```python
import numpy as np
from sklearn import preprocessing
data = np.array([[3,-1.5,2,-5.4],[0,4,-0.3,2.1],[1,3.3,-1.9,-4.3]])
```

###Remover el promedio
Usualmente es beneficioso quitar la media de cada caracteristica para que este centrada en cero esto para deshacernos de los sesgos de estas caracteristicas

```python
data_standarized = preprocessing.scale(data)
print("Mean=",data_standarized.mean(axis=0))
print("Std deviation=", data_standarized.std(axis=0))
```

###Escala
Los valores de la catracteristica pueden variar por lo que toca escalarlos para que esten en el mismo rango de valores.

```python
data_scaler = preprocessing.MinMaxScaler(feature_range=(0,1))
data_scaled = data_scaler.fit_transform(data)
print("Min max scaled data=\n", data_scaled)
```

###Normalización
Lo usamos cuando quiera ajustar los valores del vector de caracteristicas para medir en una escala en comun.

```python
data_normalized = preprocessing.normalize(data, norm ='l1')
print("L1 normalized data =\n",data_normalized)
```

###Binarización
La binarizacion se usa para convertitr el vector de caracteristicas de numerico a un vector booleano

```python
data_binarized = preprocessing.Binarizer(threshold=1.4).transform(data)
print("Binarized data =\n", data_binarized)
```

vamos a probrar con una estrategia, agregaremos datos nuevos rellenando con la media

```python
def nuevos_ejemplos_media(datos_train,n,cant_ejemplos):

  # Generar nuevos ejemplos de jugadas que son n basados en la media
  nuevos_ejemplos = pd.DataFrame({'jugada': [n] *cant_ejemplos})

  nuevos_ejemplos['palo_1'] = datos_train['palo_1'].fillna(int(datos_train[datos_train['jugada'] == n]['palo_1'].mean()))
  nuevos_ejemplos['carta_1'] = datos_train['carta_1'].fillna(int(datos_train[datos_train['jugada'] == n]['carta_1'].mean()))
  nuevos_ejemplos['palo_2'] = datos_train['palo_2'].fillna(int(datos_train[datos_train['jugada'] == n]['palo_2'].mean()))
  nuevos_ejemplos['carta_2'] = datos_train['carta_2'].fillna(int(datos_train[datos_train['jugada'] == n]['carta_2'].mean()))
  nuevos_ejemplos['palo_3'] = datos_train['palo_3'].fillna(int(datos_train[datos_train['jugada'] == n]['palo_3'].mean()))
  nuevos_ejemplos['carta_3'] = datos_train['carta_3'].fillna(int(datos_train[datos_train['jugada'] == n]['carta_3'].mean()))
  nuevos_ejemplos['palo_4'] = datos_train['palo_4'].fillna(int(datos_train[datos_train['jugada'] == n]['palo_4'].mean()))
  nuevos_ejemplos['carta_4'] = datos_train['carta_4'].fillna(int(datos_train[datos_train['jugada'] == n]['carta_4'].mean()))
  nuevos_ejemplos['palo_5'] = datos_train['palo_5'].fillna(int(datos_train[datos_train['jugada'] == n]['palo_5'].mean()))
  nuevos_ejemplos['carta_5'] = datos_train['carta_5'].fillna(int(datos_train[datos_train['jugada'] == n]['carta_5'].mean()))

  return nuevos_ejemplos
```

```python
# datos
datos_train = pd.read_csv("poker-hand-training-true.data")
datos_test = pd.read_csv("poker-hand-testing.data")

n_ejemplos_1 = nuevos_ejemplos_media(datos_train,1,2000)
n_ejemplos_2 = nuevos_ejemplos_media(datos_train,2,10000)
n_ejemplos_3 = nuevos_ejemplos_media(datos_train,3,12000)
n_ejemplos_4 = nuevos_ejemplos_media(datos_train,4,12000)
n_ejemplos_5 = nuevos_ejemplos_media(datos_train,5,12000)
n_ejemplos_6 = nuevos_ejemplos_media(datos_train,6,12000)
n_ejemplos_7 = nuevos_ejemplos_media(datos_train,7,12000)
n_ejemplos_8 = nuevos_ejemplos_media(datos_train,8,12000)
n_ejemplos_9 = nuevos_ejemplos_media(datos_train,9,12000)

datos_train = datos_train.append(n_ejemplos_1, ignore_index=True)
datos_train = datos_train.append(n_ejemplos_2, ignore_index=True)
datos_train = datos_train.append(n_ejemplos_3, ignore_index=True)
datos_train = datos_train.append(n_ejemplos_4, ignore_index=True)
datos_train = datos_train.append(n_ejemplos_5, ignore_index=True)
datos_train = datos_train.append(n_ejemplos_6, ignore_index=True)
datos_train = datos_train.append(n_ejemplos_7, ignore_index=True)
datos_train = datos_train.append(n_ejemplos_8, ignore_index=True)
datos_train = datos_train.append(n_ejemplos_9, ignore_index=True)

func_eval(datos_train,datos_test)
```
